## Cleaning Data

### Cleaning setdata_hour

```python
#agar menjaga keaslian data
data_hour = setdata_hour.copy()
```

```python
data_hour['dteday'] = pd.to_datetime(data_hour['dteday'])
```

```python
data_hour.info()
```

#### rename kolom dan value agar mudah dianalisis

```python
import pandas as pd

# Rename kolom agar lebih mudah dibaca
data_hour.rename(columns={
    'yr': 'year',
    'hr': 'hours',
    'mnth': 'month',
    'weekday': 'one_of_week',
    'weathersit': 'weather_situation',
    'windspeed': 'wind_speed',
    'cnt': 'count_cr',
    'hum': 'humidity'
}, inplace=True)

# Konversi kolom ke kategori agar lebih optimal
data_hour['month'] = data_hour['month'].astype("category")
data_hour['weather_situation'] = data_hour['weather_situation'].astype("category")
data_hour['year'] = data_hour['year'].astype("category")
data_hour['one_of_week'] = data_hour['one_of_week'].astype("category")
data_hour['season'] = data_hour['season'].astype("category")

# Mapping nilai menggunakan dictionary
month_mapping = {
    1: 'Jan', 2: 'Feb', 3: 'Mar', 4: 'Apr', 5: 'May', 6: 'Jun',
    7: 'Jul', 8: 'Aug', 9: 'Sep', 10: 'Oct', 11: 'Nov', 12: 'Dec'
}

weather_mapping = {1: 'Clear', 2: 'Misty', 3: 'Light_rainsnow', 4: 'Heavy_rainsnow'}
year_mapping = {0: '2011', 1: '2012'}
weekday_mapping = {
    0: 'Minggu', 1: 'Senin', 2: 'Selasa', 3: 'Rabu',
    4: 'Kamis', 5: 'Jumat', 6: 'Sabtu'
}
season_mapping = {1: 'Spring', 2: 'Summer', 3: 'Fall', 4: 'Winter'}

# Terapkan mapping menggunakan .map()
data_hour['month'] = data_hour['month'].map(month_mapping)
data_hour['weather_situation'] = data_hour['weather_situation'].map(weather_mapping)
data_hour['year'] = data_hour['year'].map(year_mapping)
data_hour['one_of_week'] = data_hour['one_of_week'].map(weekday_mapping)
data_hour['season'] = data_hour['season'].map(season_mapping)
```

#### mengubah value normalize temp dan humidity

```python
# Menghitung Humidity
data_hour['temp'] = round(data_hour['temp']*41, 2)
data_hour['humidity'] = data_hour['humidity']*100
```

```python
#membuat kolom untuk analisis clustering bining
def classify_humidity(humidity):
    if humidity < 45:
        return "kering"
    elif humidity >= 45 and humidity < 65:
        return "Ideal"
    else:
        return "Lembab"

data_hour["humidity_category"] = data_hour["humidity"].apply(classify_humidity)
```

```python
data_hour.head()
```

### Cleaning setdata_day

```python
#agar menjaga keaslian data
data_day = setdata_day.copy()
```

```python
data_day['dteday'] = pd.to_datetime(data_day['dteday'])
```

```python
data_day.info()
```

#### rename kolom dan value agar mudah dianalisis

```python
# Mengganti nama kolom agar lebih mudah dibaca di dataset hour_df
data_day = data_day.rename(columns={
    'yr': 'year',
    'hr': 'hours',
    'mnth': 'month',
    'weekday': 'one_of_week',
    'weathersit': 'weather_situation',
    'windspeed': 'wind_speed',
    'cnt': 'count_cr',
    'hum': 'humidity'
})

# Konversi month menjadi: 1:Jan, 2:Feb, 3:Mar, 4:Apr, 5:May, 6:Jun, 7:Jul, 8:Aug, 9:Sep, 10:Oct, 11:Nov, 12:Dec
month_mapping = {
    1: 'Jan', 2: 'Feb', 3: 'Mar', 4: 'Apr', 5: 'May', 6: 'Jun',
    7: 'Jul', 8: 'Aug', 9: 'Sep', 10: 'Oct', 11: 'Nov', 12: 'Dec'
}
data_day['month'] = data_day['month'].map(month_mapping).astype('category')

# Konversi weather_situation menjadi: 1:Clear, 2:Misty, 3:Light_RainSnow 4:Heavy_RainSnow
weather_mapping = {
    1: 'Clear', 2: 'Misty', 3: 'Light_rainsnow', 4: 'Heavy_rainsnow'
}
data_day['weather_situation'] = data_day['weather_situation'].map(weather_mapping).astype('category')

# Konversi year menjadi: 0:2011, 1:2012
year_mapping = {0: '2011', 1: '2012'}
data_day['year'] = data_day['year'].map(year_mapping).astype('category')

# Mengubah value pada kolom 'weekday' menjadi nama hari
day_mapping = {
    0: 'minggu', 1: 'senin', 2: 'selasa', 3: 'rabu',
    4: 'kamis', 5: 'jumat', 6: 'sabtu'
}
data_day['one_of_week'] = data_day['one_of_week'].map(day_mapping)
data_day['one_of_week'] = pd.Categorical(data_day['one_of_week'], categories=['minggu', 'senin', 'selasa', 'rabu', 'kamis', 'jumat', 'sabtu'], ordered=True)

# Mengubah value pada kolom 'seasonal' menjadi season
season_mapping = {
    1: 'springer', 2: 'summer', 3: 'fall', 4: 'winter'
}
data_day['season'] = data_day['season'].map(season_mapping).astype('category')
```

#### mengubah value normalize temp dan humidity menjadi nilai sebenarnya

```python
# Menghitung Humidity
data_day['temp'] = round(data_day['temp']*41, 2)
data_day['humidity'] = round(data_day['humidity']*100,2)
```

```python
def classify_humidity(humidity):
    if humidity < 45:
        return "kering"
    elif humidity >= 45 and humidity < 65:
        return "Ideal"
    else:
        return "Lembab"


data_day["humidity_category"] = data_day["humidity"].apply(classify_humidity)
```

```python
# Distribusi jumlah peminjaman sepeda per jam
plt.figure(figsize=(12, 6))
sns.histplot(setdata_hour["cnt"], bins=30, kde=True, color="#0D47A1" )
plt.title("Distribusi Jumlah Peminjaman Sepeda per Jam")
plt.xlabel("Jumlah Peminjaman")
plt.ylabel("Frekuensi")
plt.show()
```

### Insight:
Distribusi jumlah peminjaman sepeda per jam menunjukkan pola yang cenderung tidak merata, dengan beberapa jam memiliki peminjaman yang jauh lebih tinggi dibandingkan lainnya. Grafik menunjukkan bahwa sebagian besar peminjaman berada pada jumlah tertentu, sementara ada juga beberapa jam dengan peminjaman yang sangat tinggi atau sangat rendah. Pola ini bisa mengindikasikan adanya jam-jam sibuk, seperti pagi atau sore hari saat orang berangkat dan pulang kerja.

```python
# peminjaman sepeda sepanjang hari (Diagram Batang)
plt.figure(figsize=(12, 6))
sns.set_style("whitegrid")

# Menghitung rata-rata peminjaman per jam
hourly_counts = setdata_hour.groupby("hr", as_index=False)["cnt"].mean()

# Menentukan jam dengan peminjaman tertinggi
max_val = hourly_counts["cnt"].max()

# Membuat bar chart secara manual dengan warna yang ditentukan
bars = plt.bar(hourly_counts["hr"], hourly_counts["cnt"], 
               color=["#0D47A1" if cnt == max_val else "#BBDEFB" for cnt in hourly_counts["cnt"]])

# Menyesuaikan label sumbu x
plt.xticks(ticks=range(0, 24), labels=[f"{i}:00" for i in range(0, 24)], rotation=45)
plt.xlabel("Jam dalam Sehari")
plt.ylabel("Rata-rata Jumlah Peminjaman")
plt.title("Peminjaman Sepeda per Jam dalam Sehari")

plt.show()
```

### Insight:
Grafik diatas dapat disimpulkan bahwa Peminjaman sepeda cenderung rendah di pagi hari, mulai meningkat pada jam kerja (08:00 - 09:00), dan mencapai puncaknya pada sore hingga malam hari (17:00 - 19:00). Ini menunjukkan bahwa sepeda banyak digunakan untuk perjalanan kerja atau pulang kantor. Setelah jam malam, jumlah peminjaman kembali menurun.

```python
# Perbandingan peminjaman antara pengguna kasual dan terdaftar
hourly_usage = setdata_hour.groupby("hr")[["casual", "registered"]].mean()

plt.figure(figsize=(12, 6))
sns.lineplot(x=hourly_usage.index, y=hourly_usage["casual"], marker="o", label="Casual Users", color="r")
sns.lineplot(x=hourly_usage.index, y=hourly_usage["registered"], marker="o", label="Registered Users", color="blue")

plt.xticks(ticks=range(0, 24), labels=[f"{i}:00" for i in range(0, 24)], rotation=45)
plt.xlabel("Jam dalam Sehari")
plt.ylabel("Rata-rata Jumlah Peminjaman")
plt.title("Perbandingan Peminjaman Sepeda: Casual vs Registered Users")
plt.legend()
plt.show()
```

### Insight:
Grafik diatas dapat disimpulkan bahwa Pengguna terdaftar (registered) cenderung meminjam sepeda lebih banyak dibandingkan pengguna kasual (casual), terutama pada jam-jam sibuk seperti pagi dan sore hari, kemungkinan saat mereka pergi dan pulang kerja. Sementara itu, pengguna kasual lebih sering meminjam sepeda pada siang hingga sore hari, mungkin untuk rekreasi. Hal ini menunjukkan bahwa sepeda digunakan tidak hanya sebagai alat transportasi tetapi juga untuk aktivitas santai.

```python
# Pilih hanya kolom numerik
numerical_setdata_hour = setdata_hour.select_dtypes(include=np.number)

# Hitung korelasi dan tampilkan heatmap
if not numerical_setdata_hour.empty: #mengecek apakah dataframe numerik tidak kosong
    plt.figure(figsize=(12, 8))
    sns.heatmap(numerical_setdata_hour.corr(), annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
    plt.title("Korelasi Antar Variabel Numerik dalam Dataset Hourly")
    plt.show()
else:
    print("Tidak ada kolom numerik yang ditemukan.")
```

### Insight:
Grafik diatas terlihat bahwa jumlah peminjaman sepeda (cnt) memiliki korelasi positif yang kuat dengan suhu (temp) dan kelembaban rendah (humidity rendah lebih baik), tetapi kurang dipengaruhi oleh faktor seperti kecepatan angin (windspeed). Selain itu, jumlah peminjaman pengguna terdaftar (registered) lebih berkorelasi dibandingkan pengguna kasual (casual). Ini menunjukkan bahwa cuaca yang nyaman dapat meningkatkan penggunaan sepeda, terutama bagi pengguna terdaftar

```python
# Boxplot untuk melihat distribusi peminjaman berdasarkan musim
season_mapping = {1: "Musim Semi", 2: "Musim Panas", 3: "Musim Gugur", 4: "Musim Dingin"}
setdata_hour["season"] = setdata_hour["season"].map(season_mapping)

# Hitung median peminjaman untuk setiap musim
median_peminjaman = setdata_hour.groupby("season")["cnt"].median().dropna()

if not median_peminjaman.empty:  # Pastikan tidak kosong
    # Tentukan musim dengan peminjaman tertinggi
    musim_tertinggi = median_peminjaman.idxmax()
else:
    musim_tertinggi = None  # Tidak ada musim dengan peminjaman tertinggi

# Buat dictionary warna: tertinggi "#90CAF9", lainnya "#D3D3D3"
warna_palette = {season: "#90CAF9" if season == musim_tertinggi else "#D3D3D3" 
                 for season in setdata_hour["season"].unique()}

plt.figure(figsize=(12, 6))

# Terapkan warna khusus berdasarkan musim
sns.boxplot(x="season", y="cnt", hue="season", data=setdata_hour,  
            order=median_peminjaman.index if not median_peminjaman.empty else None, 
            palette=warna_palette, legend=False)  # Tambahkan hue & legend=False

plt.xlabel("Musim")
plt.ylabel("Jumlah Peminjaman")
plt.title("Distribusi Jumlah Peminjaman Sepeda Berdasarkan Musim")

plt.show()
```

### Insight:
- Musim gugur (fall) memiliki jumlah peminjaman tertinggi secara konsisten, menunjukkan bahwa kondisi cuaca di musim ini lebih mendukung aktivitas bersepeda.
- Terdapat beberapa outlier di semua musim, menandakan bahwa pada hari-hari tertentu jumlah peminjaman bisa jauh lebih tinggi atau lebih rendah dari biasanya.
- Musim berpengaruh signifikan terhadap tren peminjaman, sehingga strategi promosi atau penyediaan sepeda bisa disesuaikan berdasarkan pola musiman.

```python
# Scatter plot antara suhu dan jumlah peminjaman
plt.figure(figsize=(10, 5))
sns.scatterplot(x=setdata_hour["temp"], y=setdata_hour["cnt"], alpha=0.5, color="blue")
plt.xlabel("Suhu")
plt.ylabel("Jumlah Peminjaman")
plt.title("Hubungan Antara Suhu dan Jumlah Peminjaman Sepeda")
plt.show()
```

### Insight:
- Semakin tinggi suhu, semakin banyak jumlah peminjaman sepeda, menunjukkan bahwa cuaca hangat mendorong aktivitas bersepeda.
- Sebaran data menunjukkan korelasi positif, meskipun ada beberapa titik dengan peminjaman rendah meskipun suhu sedang ideal.
- Peminjaman tertinggi terjadi pada suhu sedang (sekitar 20-30C), menandakan kondisi optimal bagi pengguna sepeda.

```python
# Tren peminjaman sepeda berdasarkan hari dalam seminggu
plt.figure(figsize=(10, 5))

# Menambahkan hue agar palette bisa digunakan tanpa warning
sns.barplot(x="weekday", y="cnt", hue="weekday", data=setdata_hour, 
            palette=["#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#90CAF9", "#D3D3D3", "#D3D3D3"], 
            dodge=False, legend=False)
plt.xlabel("Hari dalam Seminggu")
plt.ylabel("Jumlah Peminjaman")
plt.title("Rata-rata Peminjaman Sepeda Berdasarkan Hari")

# Pastikan `weekday` berisi angka 0-6 sebelum mengganti labelnya
if setdata_hour["weekday"].nunique() == 7:
    plt.xticks(
        ticks=range(7),  # Memastikan label sesuai dengan jumlah hari
        labels=['Minggu', 'Senin', 'Selasa', 'Rabu', 'Kamis', 'Jumat', 'Sabtu']
    )

plt.show()
```

### Insight:
Peminjaman sepeda lebih tinggi pada hari kerja dibandingkan akhir pekan. Ini menunjukkan bahwa sepeda banyak digunakan untuk keperluan kerja atau sekolah. Pada hari Sabtu dan Minggu, jumlah peminjaman sedikit menurun, kemungkinan karena lebih banyak orang beristirahat atau memiliki aktivitas lain.

- **Peminjaman sepeda per jam dalam sehari** cenderung rendah di pagi hari, mulai meningkat pada jam kerja (08:00 - 09:00), dan mencapai puncaknya pada sore hingga malam hari (17:00 - 19:00). Ini menunjukkan bahwa sepeda banyak digunakan untuk perjalanan kerja atau pulang kantor. Setelah jam malam, jumlah peminjaman kembali menurun.
- **Peminjaman sepeda berdasarkan hari** menunjukkan tren lebih tinggi pada hari kerja dibandingkan akhir pekan, mengindikasikan bahwa penggunaan sepeda lebih banyak untuk aktivitas rutin sehari-hari dibandingkan rekreasi.
## Conclusion Pertanyaan 2:
Grafik diatas dapat disimpulkan bahwa Pengguna terdaftar (registered) cenderung meminjam sepeda lebih banyak dibandingkan pengguna kasual (casual), terutama pada jam-jam sibuk seperti pagi dan sore hari, kemungkinan saat mereka pergi dan pulang kerja. Sementara itu, pengguna kasual lebih sering meminjam sepeda pada siang hingga sore hari, mungkin untuk rekreasi. Hal ini menunjukkan bahwa sepeda digunakan tidak hanya sebagai alat transportasi tetapi juga untuk aktivitas santai.
